```python
# translator.translate(번역할 단어)
# 결과값 src="번역전언어", dest = "번역후언어", text= 번역된 언어 (주로 영어로 세팅되어 있음)
# # translator.translate(번역할 단어).text하면 결과만 나옴
translator = Translator()
```
